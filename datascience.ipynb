```python
# Displaying available columns/variables for ease
print("Available variables:")
print(df.columns)

# letting user choose the variable for which they want to calculate summary statistics
chosen_variable = input("Enter the variable for which you want to calculate summary statistics: ")

# Calculating summary statistics
summary_stats = {
    'Sum': df[chosen_variable].sum(),
    'Mean': df[chosen_variable].mean(),
    'Standard Deviation': df[chosen_variable].std(),
    'Skewness': df[chosen_variable].skew(),
    'Kurtosis': df[chosen_variable].kurtosis()
}

# Displaying summary statistics
print("\nSummary Statistics:")
for stat, value in summary_stats.items():
    print(f"{stat}: {value}")
```

```python
#program to calculate and show correlation of all variables:
def calculate_correlation(df):
    numeric_data = df.select_dtypes(include=['number'])

    # Calculating correlation matrix
    correlation_matrix = numeric_data.corr()

    # Showing correlation matrix
    print("Correlation of Numerical Variables:")
    print(correlation_matrix)
calculate_correlation(df)
```

# Data exploration

```python
# Calculating the count of each unique job title
job_counts = df['job_title'].value_counts()

# Sorting the values in descending order and get the top 15 jobs
top_15_jobs = job_counts.head(15)

# Converting the series to a DataFrame
top_15_jobs_df = pd.DataFrame({'Job Title': top_15_jobs.index, 'Count': top_15_jobs.values})

# Displaying the DataFrame as a table
(top_15_jobs_df)
```

```python
# Filtering the DataFrame to include only the rows corresponding to the top 15 job titles
top_15_jobs_salaries = df[df['job_title'].isin(top_15_jobs.index)]

# Calculating the mean salary for each job title
mean_salary_by_job = top_15_jobs_salaries.groupby('job_title')['salary_in_usd'].mean()

# Converting the Series to a DataFrame
mean_salary_table = pd.DataFrame({'Job Title': mean_salary_by_job.index, 'Mean Salary (USD)': mean_salary_by_job.values})

# Display the table
(mean_salary_table)
```

```python
plt.figure(figsize=(10, 6))

# Plotting the top 15 job titles with their mean salaries
bars = plt.bar(mean_salary_table['Job Title'], mean_salary_table['Mean Salary (USD)'], color=colors[:15])

# Annotating each bar with its corresponding salary amount
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 30, f"${height:.0f}",ha='center', va='bottom',  fontsize=8)

plt.title('Top 15 Job Titles with Mean Salary', fontsize=18, fontweight='bold', fontfamily='serif')
plt.xlabel("Job Title", fontsize=14, fontweight='bold', fontfamily='serif')
plt.ylabel("Mean Salary in USD", fontsize=14, fontweight='bold', fontfamily='serif')
plt.xticks(rotation=45, ha='right', fontsize=12, fontfamily='serif')
plt.yticks(fontsize=12, fontfamily='serif')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('navy')
plt.gca().spines['left'].set_color('navy')

plt.tight_layout()
plt.show()
```

```python
# Grouping the data by job title and calculating the mean salary for each job
mean_salary_by_job = df.groupby('job_title')['salary_in_usd'].mean()

# Finding the top 5 jobs with the highest mean salary
top_5_jobs = mean_salary_by_job.nlargest(5)
top_5_jobs
```

```python
# Plotting the top 5 job titles with the highest salaries
plt.figure(figsize=(10, 6))
colors = colors = ['magenta', 'khaki', 'beige', 'black', 'lightcoral']

#colors=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
#colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']


top_5_jobs.plot(kind='bar', legend=False, color=colors)
# Annotating each bar with its corresponding salary amount
for i, salary in enumerate(top_5_jobs):
    plt.text(i, salary, f"${salary:.2f}", ha='center', va='bottom', fontsize=12)

plt.title('Top 5 Job Titles with Highest Salary', fontsize=18)
plt.xlabel("Job Title", fontsize=14)
plt.ylabel("Salary in USD", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('navy')
plt.gca().spines['left'].set_color('navy')

plt.tight_layout()
plt.show()
```

```python
# Calculate the average salary for each experience level and sort the results by average salary in descending order
average_salary = df.groupby(["experience_level"])["salary_in_usd"].mean().reset_index().sort_values(by='salary_in_usd', ascending=False)

# Round the average salaries to three decimal places for clarity
round(average_salary, 3)
```

```python
# Plot the bar chart
plt.figure(figsize=(15, 6))  # Set the figure size
bars = plt.bar(average_salary["experience_level"], average_salary["salary_in_usd"], color=colors,width=0.35)

# Colors for the bars
colors = ["beige", "navy", "wheat", "pink"]


# Adding data labels to each bar
for index, value in enumerate(average_salary["salary_in_usd"]):
    plt.text(index, value, f"${value:,.0f}", ha='center', va='bottom', fontsize=13)

# Adding title and labels
plt.title("Average Salary by Experience Level",fontsize=20, fontweight='bold',fontfamily='serif')
plt.xlabel("Experience Level")
plt.ylabel("Average Salary")
plt.xticks( fontsize=12, fontfamily='serif')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('navy')
plt.gca().spines['left'].set_color('navy')


# Show the plot
plt.tight_layout()  # Adjusting layout to prevent clipping of labels
plt.show()
```
